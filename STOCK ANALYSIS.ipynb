```python
import pandas as pd
```

```python
import glob
```

```python
files = glob.glob('C:/Users/arisd/Downloads/Stock/Dataset/*')
```

```python
li = []

for file in files:
    name = file.split('\\')[1].split('.')[0]
    df = pd.read_csv(file)
    df['company'] = name
    li.append(df)
    
stock = pd.concat(li, axis=0)
```

```python
stock.head()
```

```python
stock.info()
```

```python
stock['company'].value_counts()
```

```python
stock['year'] = stock['date'].apply(lambda x:x[:4]).astype(int)
```

```python
stock.head()
```

```python
#group by year and company to find maximum close stock in each year
stock.groupby(by=['year','company']).max()['close'].unstack().fillna(0)
```

```python
#group by year and company to find maximum stock volume in each year
stock.groupby(by=['year','company']).max()['vol'].unstack().fillna(0)
```
